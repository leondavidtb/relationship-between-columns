# Insights

## Importando bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
%matplotlib inline 
```

## Lendo o arquivo

```python
df = pd.read_csv('fozcoa.csv', index_col=0, delimiter=",", low_memory=False)
```

## Convertendo colunas para Float

```python
df['PtCondensação (°C)  #10597645-x'] = pd.to_numeric(df['PtCondensação (°C)  #10597645-x'],errors='coerce')
df['Zona W2 - F4 Deslocamento (mm)'] = pd.to_numeric(df['Zona W2 - F4 Deslocamento (mm)'],errors='coerce')
df['Zona W1 - T1 Temp. (°C)'] = pd.to_numeric(df['Zona W1 - T1 Temp. (°C)'],errors='coerce')
df['Zona W1 - RH1 HR (%)'] = pd.to_numeric(df['Zona W1 - RH1 HR (%)'],errors='coerce')
df['Zona W1 - C1A Rotação (mm/m)'] = pd.to_numeric(df['Zona W1 - C1A Rotação (mm/m)'],errors='coerce')
df['Zona W1 - P1 Deslocamento (mm)'] = pd.to_numeric(df['Zona W1 - P1 Deslocamento (mm)'],errors='coerce')
df['Zona W2 - F7 Deslocamento (mm)'] = pd.to_numeric(df['Zona W2 - F7 Deslocamento (mm)'],errors='coerce')
df['Zona W2 - F8 Deslocamento (mm)'] = pd.to_numeric(df['Zona W2 - F8 Deslocamento (mm)'],errors='coerce')
df['Zona W2 - T2 Temperatura (ºC)'] = pd.to_numeric(df['Zona W2 - T2 Temperatura (ºC)'],errors='coerce')
```

## Missing values

```python
df.info()
```

```python
nLines = pd.isnull(df).any(1).to_numpy().nonzero()[0]
```

```python
nLines
```

```python
print(df.isnull().sum())
```

```python
# df.fillna(df.mean())
df.dropna(inplace=True)
```

```python
print(df.isnull().sum())
```

## Usando o K-Means

### Criando um array à partir dos dados

```python
X = np.array(df.drop(['Data', 'Hora'], axis = 1))
```

### kmeans com 3 clusters

```python
kmeans = KMeans(n_clusters=3, init = 'k-means++', random_state=0)
```

```python
kmeans.fit(X)
```

```python
df['k-classes'] = kmeans.labels_
```

```python
df['k-classes'].value_counts()
```

```python
centroids = kmeans.cluster_centers_
```

## Relação entre colunas do dataset

### Umidade relativa

Nos gráficos das umidades podemos perceber que os dados da classe 2 estão dispostos quando a umidade está acima de 60000.

Em relação à zona **W1 - T1**, podemos ver que quando a temperatura está abaixo de 20000 e a umidade abaixo dos 60000 os dados são, em sua maioria, pertencentes à classe 0. No outro caso, quando a temperatura excede 20000 os dados são, na maioria, da classe 1.

No caso da zona **W2 - T2** os dados da classe 0 estão, como no caso anterior, quando a temperatura está abaixo dos 20000. E acima diss os dados pertencem à classe 1.

Em referência à condensação, os dados da classe 0 estão no intervalo de 40000 e 60000 de umidade e condensação abaixo de 12500. E os da classe 1 estão abaixo de 60000 e condensação acima de 50000.

```python
sns.lmplot(x='Zona W1 - T1 Temp. (°C)', y='Zona W1 - RH1 HR (%)', 
           data=df, fit_reg=False, hue='k-classes')
```

```python
sns.lmplot(x='Zona W2 - T2 Temperatura (ºC)', y='Zona W1 - RH1 HR (%)', 
           data=df, fit_reg=False, hue='k-classes')
```

```python
sns.lmplot(x='PtCondensação (°C)  #10597645-x', y='Zona W1 - RH1 HR (%)', 
           data=df, fit_reg=False, hue='k-classes')
```
